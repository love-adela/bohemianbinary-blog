```python
origin = ripple_repository.get_data()
data = {
    'high': origin['high'],
    'low' : origin['low']
}
middle = {}

for x in data['high']:
    middle[x] = (data['high'][x] + data['low'][x]) / 2
data['middle'] = middle

df = pd.DataFrame(data)
df.plot()
```
